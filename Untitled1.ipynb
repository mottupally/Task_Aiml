```python
# ==============================
# Electric Vehicle Performance Analysis
# ==============================

# Import Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

# Load Dataset
df = pd.read_csv("ev_data.csv")

print(df.head())

# =====================================
# PYTHON BASICS
# =====================================

# 1. Function for energy efficiency

def calculate_efficiency(vehicle_range, battery_capacity):
    return vehicle_range / battery_capacity


df["Calculated_Efficiency"] = calculate_efficiency(
    df["Range"],
    df["Battery_Capacity"]
)

# 2. Function to classify efficiency

def classify_vehicle(efficiency):

    if efficiency > 8:
        return "Excellent"

    elif 6 <= efficiency <= 8:
        return "Good"

    elif 4 <= efficiency < 6:
        return "Average"

    else:
        return "Poor"


df["Category"] = df["Calculated_Efficiency"].apply(
    classify_vehicle
)

# 3. Filter vehicles

filtered = df[
    (df["Range"] > 400) &
    (df["Charging_Time"] < 2) &
    (df["Safety_Rating"] >= 4)
]

print("\nFiltered Vehicles:")
print(filtered)

# =====================================
# NUMPY QUESTIONS
# =====================================

battery = df["Battery_Capacity"].values
vehicle_range = df["Range"].values
price = df["Price"].values
motor = df["Motor_Power"].values
charging = df["Charging_Time"].values

print("\nMean Battery Capacity:")
print(np.mean(battery))

print("\nMean Vehicle Range:")
print(np.mean(vehicle_range))

print("\nPrice Std:")
print(np.std(price))

print("\nMaximum Motor Power:")
print(np.max(motor))

print("\nMinimum Charging Time:")
print(np.min(charging))

# Correlations

print("\nBattery vs Range:")
print(np.corrcoef(battery, vehicle_range)[0,1])

print("\nPrice vs Motor:")
print(np.corrcoef(price, motor)[0,1])

print("\nCharging vs Battery:")
print(np.corrcoef(charging, battery)[0,1])

# =====================================
# PANDAS QUESTIONS
# =====================================

print("\nMost expensive vehicle:")
print(df.loc[df["Price"].idxmax()])

print("\nCheapest vehicle:")
print(df.loc[df["Price"].idxmin()])

print("\nMaximum range vehicle:")
print(df.loc[df["Range"].idxmax()])

print("\nHighest efficiency vehicle:")
print(df.loc[df["Efficiency"].idxmax()])

# Manufacturer-wise analysis

manufacturer_stats = df.groupby(
    "Manufacturer"
)[
    ["Price","Range","Battery_Capacity"]
].mean()

print(manufacturer_stats)

# =====================================
# MATPLOTLIB
# =====================================

# Bar Chart

avg_range = df.groupby(
    "Manufacturer"
)["Range"].mean()

plt.figure(figsize=(10,5))
avg_range.plot(kind="bar")
plt.title("Average Range by Manufacturer")
plt.ylabel("Range")
plt.xticks(rotation=45)
plt.show()

# Histogram

plt.figure(figsize=(8,5))
plt.hist(df["Price"], bins=10)
plt.title("Price Distribution")
plt.xlabel("Price")
plt.show()

# Pie chart

vehicle_counts = df["Vehicle_Type"].value_counts()

plt.figure(figsize=(7,7))
plt.pie(
    vehicle_counts,
    labels=vehicle_counts.index,
    autopct='%1.1f%%'
)
plt.title("Vehicle Type Distribution")
plt.show()

# =====================================
# SEABORN
# =====================================

# Heatmap

plt.figure(figsize=(10,7))

numeric_df = df.select_dtypes(include=np.number)

sns.heatmap(
    numeric_df.corr(),
    annot=True,
    cmap='coolwarm'
)

plt.show()

# Pairplot

sns.pairplot(
    df[
        [
            "Battery_Capacity",
            "Range",
            "Motor_Power",
            "Price"
        ]
    ]
)

plt.show()

# Violin Plot

plt.figure(figsize=(8,5))

sns.violinplot(
    x="Vehicle_Type",
    y="Efficiency",
    data=df
)

plt.show()

# =====================================
# PLOTLY
# =====================================

fig = px.scatter(
    df,
    x="Battery_Capacity",
    y="Range",
    color="Manufacturer",
    size="Price",
    hover_data=["Model"],
    title="Battery Capacity vs Range"
)

fig.show()

# Dropdown Manufacturer Filter

manufacturers = df["Manufacturer"].unique()

fig = px.scatter(
    df,
    x="Battery_Capacity",
    y="Range",
    color="Manufacturer",
    size="Price"
)

buttons = []

for m in manufacturers:

    buttons.append(
        dict(
            label=m,
            method="update",
            args=[
                {
                    "visible":
                    [
                        manufacturer==m
                        for manufacturer
                        in df["Manufacturer"]
                    ]
                }
            ]
        )
    )

fig.update_layout(
    updatemenus=[
        dict(buttons=buttons)
    ]
)

fig.show()

print("Project Completed")
```
